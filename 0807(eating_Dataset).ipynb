# CGM by Stress level with 2 groups (0: Less eater, 1: more eater)

## Stress level: 0, 0.33:Slightly, 0.66:Moderately, 1: Extremely

```python
# 1. Count 계산
counts = df.groupby(['Eating_Group_Binary', 'How_stressed']).size().reset_index(name='Count')

# 2. 평균 계산
means = df.groupby(['Eating_Group_Binary', 'How_stressed'])[
    ['Closest CGM Glucose', 'CGM_30min_mean', 'CGM_60min_mean', 'CGM_90min_mean']
].mean().round(2).reset_index()

# 3. 병합
summary = pd.merge(means, counts, on=['Eating_Group_Binary', 'How_stressed'])

# 4. 레이블 붙이기
summary['Eating_Group_Label'] = summary['Eating_Group_Binary'].astype(str) + ' (' + \
                                 summary.groupby('Eating_Group_Binary')['Count'].transform('sum').astype(str) + ')'
summary['How_stressed_Label'] = summary['How_stressed'].astype(str) + ' (' + summary['Count'].astype(str) + ')'

# 5. 최종 정리
summary_final = summary[[
    'Eating_Group_Label', 'How_stressed_Label',
    'Closest CGM Glucose', 'CGM_30min_mean', 'CGM_60min_mean', 'CGM_90min_mean'
]]

# 보기 좋게 출력
summary_final
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.cm import get_cmap
from matplotlib import colormaps



# 1. Melt to long format
plot_data = summary_final.melt(
    id_vars=['Eating_Group_Label', 'How_stressed_Label'],
    var_name='Time', value_name='Glucose'
)

# 2. Extract group number (0 or 1)
plot_data['Group_Num'] = plot_data['Eating_Group_Label'].str.extract(r'(\d)').astype(str)

# 3. Extract stress level (for sorting)
plot_data['Stress_Level'] = plot_data['How_stressed_Label'].str.extract(r'([0-9.]+)').astype(float)

# 4. Set Time order
time_order = ['Closest CGM Glucose', 'CGM_30min_mean', 'CGM_60min_mean', 'CGM_90min_mean']
time_labels = ['0 min', '30 min', '60 min', '90 min']
time_map = dict(zip(time_order, time_labels))

plot_data['Time'] = plot_data['Time'].map(time_map)
plot_data['Time'] = pd.Categorical(plot_data['Time'], categories=time_labels, ordered=True)

# 5. Sort for consistent plotting
plot_data = plot_data.sort_values(by=['Group_Num', 'Stress_Level', 'Time'])

# 6. Color maps
less_colors = colormaps['Blues']([0.3, 0.5, 0.7, 0.9])
more_colors = colormaps['Reds']([0.3, 0.5, 0.7, 0.9])
stress_levels = sorted(plot_data['Stress_Level'].unique())

# 7. Mapping color per group × stress
color_map = {}
for i, level in enumerate(stress_levels):
    color_map[('0', level)] = less_colors[i]
    color_map[('1', level)] = more_colors[i]

# 8. Marker and line style
marker_map = {'0': 'o', '1': 'x'}
linestyle_map = {'0': '-', '1': '--'}
facecolor_map = {'0': 'none', '1': None}

# Stress level mapping
stress_label_map = {
    0.0: 'Not at all',
    0.3333333333333333: 'Slightly',
    0.6666666666666666: 'Moderately',
    1.0: 'Extremely'
}

# Plot
plt.figure(figsize=(10, 6))
for (group, level), group_df in plot_data.groupby(['Group_Num', 'Stress_Level']):
    group_name = 'Less eater' if group == '0' else 'More eater'
    stress_name = stress_label_map.get(level, str(level))
    label = f"{group_name} - {stress_name}"

    plt.plot(
        group_df['Time'], group_df['Glucose'],
        marker=marker_map[group],
        linestyle=linestyle_map[group],
        markerfacecolor=facecolor_map[group],
        markeredgecolor='black',
        color=color_map[(group, level)],
        label=label
    )

plt.title('Glucose Over Time by Stress Level and Eating Group')
plt.xlabel('Time After EMA')
plt.ylabel('Glucose (mg/dL)')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.legend(title='Group - Stress Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
```

# Binary by stress or not (4 stress level -> 2 level)

```python
# 1. 이진 스트레스 그룹 생성 (0: Not at all, 1: Any stress)
df['Stress_Binary'] = (df['How_stressed'] > 0.0).astype(int)

# 2. delta_90min 계산
df['delta_90min'] = df['CGM_90min_mean'] - df['Closest CGM Glucose']

# 3. 그룹별 요약 통계 계산
delta_summary_bin = df.groupby(['Eating_Group_Binary', 'Stress_Binary'])['delta_90min'].agg(['mean', 'std', 'count']).round(2).reset_index()

# 4. 레이블 추가
delta_summary_bin['Eating_Group_Label'] = delta_summary_bin['Eating_Group_Binary'].map({0: 'Less eater', 1: 'More eater'})
delta_summary_bin['Stress_Label'] = delta_summary_bin['Stress_Binary'].map({0: 'Not at all', 1: 'Any stress'})

# 5. 최종 테이블 정리
delta_summary_bin = delta_summary_bin[[
    'Eating_Group_Label', 'Stress_Label', 'mean', 'std', 'count'
]].rename(columns={
    'mean': 'Δ Glucose (0→90min) Mean',
    'std': 'SD',
    'count': 'n'
})

# 6. 출력
delta_summary_bin
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Stress_Binary 생성
df['Stress_Binary'] = (df['How_stressed'] > 0.0).astype(int)

# 2. 평균 계산
glucose_binary_summary = df.groupby(['Eating_Group_Binary', 'Stress_Binary'])[
    ['Closest CGM Glucose', 'CGM_30min_mean', 'CGM_60min_mean', 'CGM_90min_mean']
].mean().reset_index()

# 3. Melt to long format
plot_data = glucose_binary_summary.melt(
    id_vars=['Eating_Group_Binary', 'Stress_Binary'],
    var_name='Time', value_name='Glucose'
)

# 4. Time 순서 정렬
time_order = ['Closest CGM Glucose', 'CGM_30min_mean', 'CGM_60min_mean', 'CGM_90min_mean']
time_label_map = {
    'Closest CGM Glucose': '0 min',
    'CGM_30min_mean': '30 min',
    'CGM_60min_mean': '60 min',
    'CGM_90min_mean': '90 min'
}
plot_data['Time'] = pd.Categorical(plot_data['Time'], categories=time_order, ordered=True)
plot_data['Time_Label'] = plot_data['Time'].map(time_label_map)

# 5. 그룹 이름 지정
plot_data['Eating_Group'] = plot_data['Eating_Group_Binary'].map({0: 'Less eater', 1: 'More eater'})
plot_data['Stress_Level'] = plot_data['Stress_Binary'].map({0: 'Not at all', 1: 'Any stress'})

# 6. Plot
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=plot_data,
    x='Time_Label', y='Glucose',
    hue='Stress_Level', style='Eating_Group',
    markers=True, dashes=True
)

plt.title('Glucose Over Time by Binary Stress Level and Eating Group')
plt.xlabel('Time After EMA')
plt.ylabel('Glucose (mg/dL)')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.legend(title='Stress Level × Eating Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
